```python
df['State'] = df['State'].str.strip()

state_mapping={
    'Assam':1, 'Karnataka':2, 'Kerala':3, 'Meghalaya':4, 'West Bengal':5, 'Puducherry':6, 'Goa':7,
 'Andhra Pradesh':8, 'Tamil Nadu':9, 'Odisha':10, 'Bihar':11, 'Gujarat':12, 'Madhya Pradesh':13,
 'Maharashtra':14, 'Mizoram':15, 'Punjab':16, 'Uttar Pradesh':17, 'Haryana':18,
 'Himachal Pradesh':19, 'Tripura':20, 'Nagaland':21, 'Chhattisgarh':22, 'Uttarakhand':23,
 'Jharkhand':24, 'Delhi':25, 'Manipur':26, 'Jammu and Kashmir':27, 'Telangana':28,
 'Arunachal Pradesh':29, 'Sikkim':30
}

df['State'] = df['State'].map(state_mapping).fillna(0)

df['State'] = df['State'].astype(int)

print(df['State'])
```

```python
x=df.drop('Yield',axis=1)
y=df['Yield']
```

```python
from sklearn.model_selection  import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.neighbors import KNeighborsRegressor

from sklearn.metrics import mean_absolute_error,r2_score
```

```python
models={
    'Decision':DecisionTreeRegressor(),
    'LinearRegression':LinearRegression(),
    'Lasso':Lasso(),
    'Ridge':Ridge(),
    'KNN':KNeighborsRegressor()
}
```

```python
for name,md in models.items():
    md.fit(x_train,y_train)
    y0=md.predict(x_test)
    print(f"mean_absolute_error: {name} ",mean_absolute_error(y_test,y0))
    print(f"r2_score : {name}" ,r2_score(y_test,y0))
```

```python
model=DecisionTreeRegressor()
```

```python
model.fit(x_train,y_train)
```

```python
y2_pred=model.predict(x_test)

training_data_accuracy = mean_absolute_error(y_test,y2_pred)
training_r2_score = r2_score(y_test,y2_pred)
print("Training Data Mean Squared Error:", training_data_accuracy)
print("r_2 score:", training_r2_score)
```
